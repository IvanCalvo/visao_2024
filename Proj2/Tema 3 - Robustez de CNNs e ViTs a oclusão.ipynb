Geração da oclusão

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = np.array(Image.open('data/oxford_pets/images/basset_hound_53.jpg'))

# Extração de uma região da imagem
bbox = [100, 100, 300, 300]
patch = img[bbox[0]:bbox[2], bbox[1]:bbox[3]]
# Média de intensidade
color = np.mean(patch)
print(color)
# Substituição dos valores
img[bbox[0]:bbox[2], bbox[1]:bbox[3]] = color

plt.imshow(img)
```
